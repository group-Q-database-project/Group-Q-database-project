```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#chart1 = pd.read_csv('C:/Users/Hibak Ramadan/Downloads/Chart_1.csv')
url = 'https://raw.githubusercontent.com/group-Q-database-project/Group-Q-database-project/refs/heads/main/Chart_1.csv'
chart1 = pd.read_csv(url)
# Average salary by city (Junior roles)
chart1['city_name'] = chart1['city_name'].str.lower()

# Chart 1: Average salary by city (Junior roles)
plt.figure(figsize=(10, 6))
plt.bar(chart1['city_name'], chart1['average_salary'], color='blue')
plt.xlabel('City Name', fontsize=12)
plt.ylabel('Average Salary (EUR)', fontsize=12)
plt.title('Average Salary by City for Junior Roles', fontsize=14)
plt.xticks(rotation=45, fontsize=7)
plt.grid(axis='y', alpha=0.5)
plt.tight_layout()
plt.savefig('average_salary_by_city.png', dpi=300, bbox_inches='tight')
plt.show()

```

```python
url2 = 'https://raw.githubusercontent.com/group-Q-database-project/Group-Q-database-project/refs/heads/main/Chart_2.csv'
data = pd.read_csv(url2)

top_positions = data.sort_values(by="average_salary", ascending=False).head(5)

plt.figure(figsize=(8, 8))
plt.pie(
    top_positions["average_salary"],
    labels=top_positions["position_name"],
    autopct="%1.1f%%",
    startangle=140,
    colors=plt.cm.Paired.colors
)

plt.legend(
    top_positions["position_name"], 
    title="Position Name", 
    loc="best", 
    fontsize=7
)

plt.title("Proportion of Average Salary for Top Positions", fontsize=14)
plt.tight_layout()
plt.show()

```

```python
url3= 'https://raw.githubusercontent.com/group-Q-database-project/Group-Q-database-project/refs/heads/main/Chart_3.csv'
chart3 = pd.read_csv(url3)
average_salary_by_city = chart3.groupby('city_name')['average_salary'].mean()

average_salary_by_city_sorted = average_salary_by_city.sort_values(ascending=False)

plt.figure(figsize=(10, 6))
sns.barplot(x=average_salary_by_city_sorted.index, y=average_salary_by_city_sorted.values, palette="viridis")

# Add labels and title
plt.title('Average Salary by City for Females', fontsize=16)
plt.xlabel('City', fontsize=12)
plt.ylabel('Average Salary', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Show the plot
plt.tight_layout()
plt.show()
```
